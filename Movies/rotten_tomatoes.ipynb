```python
def data_investigate(df):
    print('No of Rows : ', df.shape[0])
    print('No of Coloums : ', df.shape[1])
    print('**'*25)
    print('Colums Names : \n', df.columns)
    print('**'*25)
    print('Datatype of Columns : \n', df.dtypes)
    print('**'*25)
    print('Missing Values : ')
    c = df.isnull().sum()
    c = c[c > 0]
    print(c)
    print('**'*25)
    print('Missing vaules %age wise :\n')
    print((100*(df.isnull().sum()/len(df.index))))
    print('**'*25)
    print('Pictorial Representation : ')
    plt.figure(figsize = (5, 5))
    sns.heatmap(df.isnull(), yticklabels = False, cbar = False)
    plt.show()
```

```python
i = 0
count_id = 0
count_roto = 0

m_rotorating = []

while(i<len(df.index)):
    print(i)
    val_type = df['Type'].values[i]
    if(val_type==0):
        val_movie = df['Title'].values[i]
        movies = roto.search(term=val_movie, limit=1)
        movie_rotorating = ""
        c1=0
        try:
            for r in movies['movies']:
                if(c1==0):
                    movie_rotorating = r['meterScore']
                    print(r['name'], ' : ', r['meterScore'])
                
                else:
                    movie_rotorating = movie_rotorating + "," + r['meterScore']
                c1+=1
            m_rotorating.append(movie_rotorating)
        except:
            m_rotorating.append("NA")
    else:
        pass
    i = i + 1
```
